# Code for generating combinations using csv files
##Naming convention

"forged-audio folder" means - the audios do not contain any sound. The file name starts with the following information:
    
(Note: The starting "0" and "1" for the label information. "a" and "g" are for audio and geophone respectively. "1", "2",....,"700" are number of audios and geophone samples.)

*   "0_a_1_........." 
*   "0_a_2_........." and so on


"genuine-audio folder" means - the audios contain the sound of human movement

*   "1_a_1_........."
*   "1_a_2_........." and so on

"forged-geophone folder" means - geophones signal contains no movement information

*   "0_g_1_........."
*   "0_g_2_........." and so on

"genuine-geophone folder" means - geophone signal contains humam movement information

*    "1_g_1_........."
*    "1_g_2_........." and so on

## Following script is then used to generate combinations of audio and geophone data to be used in the multimodal Siamese neural network

Thi scrip will only generate labels of the audio and geophone files in csv file information.

Please change the path information to use it according to your needs.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import csv as c
```

For Training sets

```python
acce_files = pd.read_csv('/content/drive/MyDrive/har-sense/train/train-acce.csv', header=None)
gyro_files = pd.read_csv('/content/drive/MyDrive/har-sense/train/train-gyro.csv', header=None)
```

For Validation sets

```python
acce_files = pd.read_csv('/content/drive/MyDrive/har-sense/valid/valid-acce.csv', header=None)
gyro_files = pd.read_csv('/content/drive/MyDrive/har-sense/valid/valid-gyro.csv', header=None)
```

```python
acce_a=[]
acce_b=[]
acce_y=[]
#j = range(len(audio_files))
#print(j)
for i in acce_files.index:
  a_path = acce_files.loc[i][0]
  #print(audio_files.loc[i][0])
  for j in acce_files.index:
    b_path=acce_files.loc[j][0]
    acce_a.append(a_path)
    acce_b.append(b_path)
    if(a_path[0]==b_path[0]):
      acce_y.append(ord(a_path[0])-48)
    else:
      acce_y.append(-1)
```
